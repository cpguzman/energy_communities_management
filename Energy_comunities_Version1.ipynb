```python
def rule_v2g_status(model, v2g, time):
    return model.Xv2gDischarge_v_t[v2g, time] + model.Xv2gCharge_v_t[v2g, time] <= 1

model.v2g_status_constraint = pyo.Constraint(model.V2G_Set, model.SetTimeIntervals, rule=rule_v2g_status)
```

```python
def rule_Inicial_Energy_V2G(model, v2g, time):
  if model.Ev2gInitial_v_t[v2g, time] != 0:
    return model.VarEnergyV2G_v_t[v2g, time] == model.Ev2gInitial_v_t[v2g, time]
  else:
    return pyo.Constraint.Skip
model.EnergyInitial_V2G_constraint = pyo.Constraint(model.V2G_Set, model.SetTimeIntervals, rule=rule_Inicial_Energy_V2G)
```

```python
def rule_rolling_Energy_V2G(model, v2g, time):
  if model.Ev2gInitial_v_t[v2g, time] == 0 and model.XconexionStatus_v_t[v2g, time] == 1:
    return model.VarEnergyV2G_v_t[v2g, time] == model.VarEnergyV2G_v_t[v2g, model.SetTimeIntervals.prev(time)] + model.VarPv2gCharge_v_t[v2g, time] - model.VarPv2gDischarge_v_t[v2g, time]
  else:
    return pyo.Constraint.Skip
model.rolling_Energy_V2G_constraint = pyo.Constraint(model.V2G_Set, model.SetTimeIntervals, rule=rule_rolling_Energy_V2G)
```

```python
def rule_ConexionStatus_Energy_V2G(model, v2g, time):
  if model.XconexionStatus_v_t[v2g, time] == 0:
    return model.VarEnergyV2G_v_t[v2g, time] == 0
  else:
    return pyo.Constraint.Skip
model.ConexionStatus_Energy_V2G_constraint = pyo.Constraint(model.V2G_Set, model.SetTimeIntervals, rule=rule_ConexionStatus_Energy_V2G)
```

```python
def rule_Exit_Energy_V2G(model, v2g, time):
  if model.ParamEv2gExitRequired_v_t[v2g, time] != 0 and model.XconexionStatus_v_t[v2g, time] == 1:
    return model.VarEnergyV2G_v_t[v2g, time] >= model.ParamEv2gExitRequired_v_t[v2g, time]
  else:
    return pyo.Constraint.Skip
model.Exit_Energy_V2G_constraint = pyo.Constraint(model.V2G_Set, model.SetTimeIntervals, rule=rule_Exit_Energy_V2G)
```

```python
def rule_Minimum_Energy_V2G(model, v2g, time):
  if model.XconexionStatus_v_t[v2g, time] == 1:
    return model.VarEnergyV2G_v_t[v2g, time] >= model.Ev2gMax_v[v2g]*V2G_Characteritics_dict["11 Minimun Technical SOC (%)"][v2g]/100
  else:
    return pyo.Constraint.Skip
model.Minimum_Energy_V2G_constraint = pyo.Constraint(model.V2G_Set, model.SetTimeIntervals, rule=rule_Minimum_Energy_V2G)
```

```python
def rule_Pv2gDischarge(model, v2g, time):
  return model.VarPv2gDischarge_v_t[v2g, time] <= model.Pv2gDischargeMax_v[v2g] * model.XconexionStatus_v_t[v2g, time] * model.Xv2gDischarge_v_t[v2g, time]
model.Pv2gDischarge_constraint = pyo.Constraint(model.V2G_Set, model.SetTimeIntervals, rule=rule_Pv2gDischarge)
```

```python
def rule_Pv2gCharge(model, v2g, time):
  return model.VarPv2gCharge_v_t[v2g, time] <= model.Pv2gChargeMax_v[v2g] * model.XconexionStatus_v_t[v2g, time] * model.Xv2gCharge_v_t[v2g, time]
model.Pv2gCharge_constraint = pyo.Constraint(model.V2G_Set, model.SetTimeIntervals, rule=rule_Pv2gCharge)
```

```python
def rule_battery_status(model, storage, time):
    return model.XstorageDischarge_b_t[storage, time] + model.XstorageCharge_b_t[storage, time] <= 1

model.battery_status_constraint = pyo.Constraint(model.BatterySet, model.SetTimeIntervals, rule=rule_battery_status)
```

```python
def rule_PstorageCharge(model, storage, time):
  return model.VarPstorageCharge_b_t[storage, time] <= model.ParamPstorageChargeLimit_b_t[storage, time] * model.XstorageCharge_b_t[storage, time]
model.PstorageCharge_constraint = pyo.Constraint(model.BatterySet, model.SetTimeIntervals, rule=rule_PstorageCharge)
```

```python
def rule_PstorageDischarge(model, storage, time):
  return model.VarPstorageDischarge_b_t[storage, time] <= model.ParamPstorageDischargeLimit_b_t[storage, time] * model.XstorageDischarge_b_t[storage, time]
model.PstorageDischarge_constraint = pyo.Constraint(model.BatterySet, model.SetTimeIntervals, rule=rule_PstorageDischarge)
```

```python
def rule_Inicial_Energy_Storage(model, storage, time):
  ChargeEfficiency = StorageCharacteritics_dict["8 Charge Efficiency (%)"][storage] / 100
  DischargeEfficiency = 100 / StorageCharacteritics_dict["9 Discharge Efficiency (%)"][storage]
  if time == 't_0':
    SoC = StorageCharacteritics_dict["10 Initial State (%)"][storage]/100
    return model.VarEnergyStorage_b_t[storage, time] == model.EstorageMax_b[storage] * SoC + (model.VarPstorageCharge_b_t[storage, time] * ChargeEfficiency) - (model.VarPstorageDischarge_b_t[storage, time] * DischargeEfficiency)
  else:
    return model.VarEnergyStorage_b_t[storage, time] == model.VarEnergyStorage_b_t[storage, model.SetTimeIntervals.prev(time)] + (model.VarPstorageCharge_b_t[storage, time] * ChargeEfficiency) - (model.VarPstorageDischarge_b_t[storage, time] * DischargeEfficiency)

model.Inicial_Energy_Storage_constraint = pyo.Constraint(model.BatterySet, model.SetTimeIntervals, rule=rule_Inicial_Energy_Storage)
```

```python
def rule_SolarGenertionPriority(model, gen, time):
  if GenCharacteritics_dict["2 Generator Type"][gen] == 2:
    return model.VarPgen_g_t[gen, time]>= model.ParamPGenForecast_g_t[gen, time]
  else:
    return pyo.Constraint.Skip
model.SolarGenertionPriority = pyo.Constraint(model.GenSet, model.SetTimeIntervals, rule=rule_SolarGenertionPriority)
```

```python
def rule_Contract_TypeI(model, gen, time):
  if GenCharacteritics_dict["5 Type of Contract"][gen] == 1:
    return model.VarPgen_g_t[gen, time] <= model.ParamPGenForecast_g_t[gen, time]
  else:
    return pyo.Constraint.Skip

model.Contract_TypeI_constraint = pyo.Constraint(model.GenSet, model.SetTimeIntervals, rule=rule_Contract_TypeI)
```

```python
def rule_Contract_TypeII(model, gen, time):
  if GenCharacteritics_dict["5 Type of Contract"][gen] == 2:
    return model.VarPgenExcess_g_t[gen, time] + model.VarPgen_g_t[gen, time] == model.ParamPGenForecast_g_t[gen, time]
  else:
    return pyo.Constraint.Skip

model.Contract_TypeII_constraint = pyo.Constraint(model.GenSet, model.SetTimeIntervals, rule=rule_Contract_TypeII)
```

```python
def rule_balance_power(model, time):
    total_generation = sum(model.VarPgen_g_t[gen, time] + model.VarPgenExcess_g_t[gen, time] for gen in model.GenSet)
    total_load = sum(model.ParamPloadForecast_l_t[load, time] for load in model.LoadSet)
    toal_Battery_Power = sum(model.VarPstorageCharge_b_t[storage, time] - model.VarPstorageDischarge_b_t[storage, time] for storage in model.BatterySet)
    toal_V2G_Power = sum(model.VarPv2gCharge_v_t[v2g, time] - model.VarPv2gDischarge_v_t[v2g, time] for v2g in model.V2G_Set)
    return total_generation + model.VarPimp_t[time] - model.VarPexp_t[time] - toal_Battery_Power - toal_V2G_Power == total_load

model.balance_power_constraint = pyo.Constraint(model.SetTimeIntervals, rule=rule_balance_power)
```

```python
def objective_rule(model):
  CostoPowerExport = sum(model.VarPexp_t[time] * model.Param_C_ExpPrice_t[time] for time in model.SetTimeIntervals)
  CostoPowerImport = sum(model.VarPimp_t[time] * model.Param_C_ImpPrice_t[time] for time in model.SetTimeIntervals)
  CostoGenPower = sum(model.VarPgen_g_t[gen, time] * model.ParamPGenCost_g_t[gen, time] for gen in model.GenSet for time in model.SetTimeIntervals)
  CostoGenPowerExcess = sum(model.VarPgenExcess_g_t[gen, time] * model.ParamPGenExcessCost_g_t[gen, time] for gen in model.GenSet for time in model.SetTimeIntervals)
  CostoBattery = sum((model.VarPstorageDischarge_b_t[storage, time] * PstorageDischargePrice_dict[time][storage]) - (model.VarPstorageCharge_b_t[storage, time] * PstorageChargePrice_dict[time][storage] ) for storage in model.BatterySet for time in model.SetTimeIntervals)
  CostoV2G = sum((model.VarPv2gDischarge_v_t[v2g, time] * V2G_DischargePrice_dict[time][v2g]) - (model.VarPv2gCharge_v_t[v2g, time] * V2G_ChargePrice_dict[time][v2g] ) for v2g in model.V2G_Set for time in model.SetTimeIntervals)
  return CostoPowerImport + CostoGenPower + CostoGenPowerExcess + CostoBattery - CostoPowerExport

model.Objective = pyo.Objective(rule=objective_rule, sense=pyo.minimize)
```
